```python
# import library yang sekiranya dibutuhkan
import pandas as pd
import numpy as np
import nltk
import string 
import re
```

```python
# masukan data kedalam project
dataclean = pd.read_csv("Downloads/testing.csv", sep=";")
```

```python
dataclean
```

```python
# CLEANSING
# drop kolom description
cleansing=dataclean.drop(["Description"], axis = 1)
```

```python
# drop row dengan nilai null/NA
cleansing=cleansing.dropna(axis = 0)
```

```python
# ambil row dengan ketentuan nilai kolom Quantity > 0
cleansing = cleansing[cleansing.Quantity > 0]
```

```python
cleansing
```

```python
# ambil row dengan ketentuan nilai kolom Price != 0
cleansing = cleansing[cleansing.Price != 0]
cleansing
```

```python
# merubah tipe data customer id mendaji INT
cleansing["Customer ID"] = cleansing["Customer ID"].astype("int")

# membuat kolom berisi perkalian antara kolom Quantity dan Price sebagai Total
cleansing["Total"]=(cleansing.Quantity*cleansing.Price).round(2)
cleansing
```

```python
# import data menjadi csv
cleansing.to_csv('clean_test.csv', index=False)
```
